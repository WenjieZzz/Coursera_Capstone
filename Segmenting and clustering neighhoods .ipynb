# 1. Data Preparation - Scraping

```python
import pandas as pd
import numpy as np
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
df = pd.read_html(url)[0]
df
```

## Drop the rows with 'Not assigned' Borough

```python
df = df[df["Borough"] != 'Not assigned']
df
```

## If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough. 

```python
for index, row in df.iterrows():
    if row["Neighbourhood"] == 'Not assigned':
        row["Neighbourhood"] = row["Borough"]
df
```

## Combine two rows' Neighbourhood, seperated with a comma, if they have the same Borough

```python
grouped = df.groupby(['Postcode','Borough']).agg(', '.join)
df = grouped.reset_index()
df
```

# 2. Get the latitude and the longitude coordinates

## Read the geographical coordinates

```python
geo = pd.read_csv("https://cocl.us/Geospatial_data")
geo.head()
```

```python
geodf = pd.merge(df,geo,how = 'left',left_on = 'Postcode', right_on = "Postal Code" )
geodf.head()
```

## Drop the Postal Code column

```python
geodf = geodf.drop('Postal Code', axis = 1)
geodf.head()
```
